```python
# this code is doing text-to-video generation using the Hugging Face Inference API (specifically the “Wan 2.2” video generation model).

import os
from huggingface_hub import InferenceClient
from IPython.display import Video
import random

# Set Hugging Face token

os.environ['HF_TOKEN'] = '' #Add your hugging face token
```

```python

# Create client
client = InferenceClient(
    api_key=os.environ['HF_TOKEN'],
    provider="auto"
)

prompts = [
   
    "A shiny black car riding on the streets of Dubai"
]
```

```python

for i, prompt in enumerate(prompts):
    # Random seed for unique video
    seed = random.randint(1, 999999)

    video_bytes = client.text_to_video(
        prompt,
        model="Wan-AI/Wan2.2-T2V-A14B",
        seed=seed   # ✅ ensures different video for each prompt
    )

    video_path = f"/content/wan_video_{i}.mp4"
    with open(video_path, "wb") as f:
        f.write(video_bytes)

    print(f"Prompt: {prompt} ✅ Video saved at {video_path}")
    display(Video(video_path, embed=True))
```
